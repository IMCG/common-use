# dump_data

Dump data from binary data, inpur paras are file name, start position and length.
Output will be in hex format.

```python
import numpy as np
import pandas as pd
import csv
import sys
import os

def dump_data(dump_file, dump_start, dump_length):

    dump_file_size = os.path.getsize(dump_file)

    if dump_start <= 0 or dump_start>dump_file_size:
        raise ValueError('Invalid start')

    if dump_start+dump_length > dump_file_size:
        raise ValueError('Invalid length')

    # Create a dtype with the binary data format and the desired column names
    dt = np.dtype([('data1', 'uint32'), ('data2', 'uint32'), ('data3', 'uint32'), ('data4', 'uint32')])
    file = open(dump_file, 'rb')
    data = np.fromfile(file, dtype=dt)
    df = pd.DataFrame(data.tolist(), columns=data.dtype.names)

    output_file = open('dump_output.txt', 'w')
    for i in range(dump_start/16, (dump_start+dump_length)/16):
        print '0x%0.8x: 0x%0.8x 0x%0.8x 0x%0.8x 0x%0.8x' % (i*16, df.iloc[i]['data1'], df.iloc[i]['data2'], df.iloc[i]['data3'], df.iloc[i]['data4'])
        output_file.writelines('0x%0.8x: 0x%0.8x 0x%0.8x 0x%0.8x 0x%0.8x\n' % (i*16, df.iloc[i]['data1'], df.iloc[i]['data2'], df.iloc[i]['data3'], df.iloc[i]['data4']))
    output_file.close()


if __name__ == '__main__':
    if sys.argv[0] == 'dump_data.py':
        dump_file = sys.argv[1]
        dump_start = int(sys.argv[2], 16)
        dump_length = int(sys.argv[3], 16)
    else:
        dump_file = 'cp_data.bin'
        dump_start = 0x200
        dump_length = 0x800

    print 'Dump:', dump_file, 'from', dump_start, ',length', dump_length, '\n'
    dump_data(dump_file, dump_start, dump_length)
```
